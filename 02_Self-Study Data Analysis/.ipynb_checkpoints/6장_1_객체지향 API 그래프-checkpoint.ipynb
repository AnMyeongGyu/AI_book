```python
import pandas as pd

ns_book7 = pd.read_csv('ns_book7.csv', low_memory=False)
ns_book7.head()
```

```python
#     30 
top30_pubs = ns_book7[''].value_counts()[:30]
top30_pubs
```

```python
#    30   boolean 
top30_pubs_idx = ns_book7[''].isin(top30_pubs.index)
top30_pubs_idx
```

```python
#   30  
top30_pubs_idx.sum()
```

```python
# sample 1000  
ns_book8 = ns_book7[top30_pubs_idx].sample(1000, random_state=42)
ns_book8.head()
```

```python
fig, ax = plt.subplots(figsize=(10, 8))
ax.scatter(ns_book8[''], ns_book8[''])
ax.set_title(' ')
```

```python
plt.rcParams['lines.markersize']
```

```python
# marker    
fig, ax = plt.subplots(figsize=(10, 8))
ax.scatter(ns_book8[''], ns_book8[''], s=ns_book8[''])
ax.set_title(' ')
fig.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 8))
ax.scatter(ns_book8[''], ns_book8[''], 
           linewidths=0.5, edgecolors='k', alpha=0.3,
           s=ns_book8['']*2, c=ns_book8[''])
ax.set_title(' ')
fig.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 8))
sc = ax.scatter(ns_book8[''], ns_book8[''], 
                linewidths=0.5, edgecolors='k', alpha=0.3,
                #cmap = 'jet'  
                s=ns_book8['']**1.3, c=ns_book8[''], cmap='jet')
ax.set_title(' ')
fig.colorbar(sc)
fig.show()
```
